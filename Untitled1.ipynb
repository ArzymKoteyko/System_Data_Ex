```python
class System_constructor:
    
    system_id = None
    system_name = None
    system_information = None
    tree = None
    urls = None
    
    
    
    
    def __init__(self, system_name):
        
        self.system_name = system_name
        self.system_id   = self.get_id()
        
        
        
        
    def get_id(self):
        
        r = requests.get('https://www.edsm.net/api-v1/systems?systemName={}&showId=1'.format(self.system_name))
        system_id = r.json()[0]['id']
        return system_id
    
    
    
    
    def get_information(self):
        
        request = requests.get('https://www.edsm.net/en/system/bodies/id/{}/name/{}'.format(self.system_id, self.system_name.replace(' ', '+')))
        information = str(request.content)
        information = information.replace('\\n', '')
        information = information.replace('\\r', '')
        information = information.replace('\\t', '')
        information = information.replace(' /', '')
        information = information.replace('    ', '')
        
        self.system_information = information
    
    
    
    
    def find_between_tags(self, content, tag, attribute = ''):
        
        if attribute != '':
            start_tag  = '<' + tag + ' ' + attribute
        else:
            start_tag  = '<' + tag
            
        finish_tag = '</' + tag + '>'

        open_tag   = '<' + tag
        close_tag  = '</' + tag + '>'
        
        start = content.find(start_tag)
        
        result = []
        
        if start == -1:
            return result
        
        else:
            while start != -1:
        
                cursor = start
                simbol = content[cursor]

                while simbol != '>':
                    cursor += 1
                    simbol = content[cursor]
                cursor += 1

                content_start = cursor
                counter = 1
                while counter != 0:
                    if content[cursor:cursor+len(open_tag)] == open_tag:
                        counter += 1
                    elif content[cursor:cursor+len(close_tag)] == close_tag:
                        counter -= 1
                    cursor += 1
                content_end = cursor - 1

                result.append(content[content_start:content_end])
                content = content[0:start] + content[content_end + len(close_tag):len(content)]
                start = content.find(start_tag)
        return result
    
    
    
    
    def find_attribute(self, content, attribute):
        
        start = content.find(attribute)
        Res = []
        
        if start == -1:
            return Res  
        
        while start != -1:
            
            start += len(attribute) + 2 
            cursor = start
            c = 1
        
            while c != 0:
                if content[cursor] == '"':
                    c -= 1
                cursor += 1
            finish = cursor - 1

            Res.append(content[start:finish])
            content = content[finish::]
            start = content.find(attribute)
            
        return Res
    
    
    
    
    def find_name(self, content):
        
        result = self.find_between_tags(content, 'a', 'href="')
        if len(result) != 0:
            return self.find_between_tags(content, 'a', 'href="')[0]
        else:
            return self.find_between_tags(content, 'div', 'class="card-header"')[0]
    
    
    
    
    def find_url(self, content):
        
        start = content.find('<a ')
    
        i = start
        while content[i] != '>':
            i += 1
        
        finish = i - 1
        start += 9
    
        return 'https://www.edsm.net{}'.format(content[start:finish])


    
    
    def alocate_tree(self, content):
        
        items = self.find_between_tags(content, 'div', 'class="container mb-4"')
        full_tree = {}
        full_urls = {}
        for i in items:
            tree, urls = self.construct_tree(i)
            full_tree |= tree
            full_urls |= urls
            
        self.tree = full_tree
        self.urls = full_urls
            
        
    def construct_tree(self, content, tree = {}, urls = {}):
    
        try:
            information = self.find_between_tags(content, 'div', 'class="card mb-0"')[0]
        except:
            return None
        name = self.find_name(information)
        
        url = self.find_url(information)
        urls[name] = url
        
        children = self.find_between_tags(content, 'div', 'class="child"')
        
        if len(children) != 0:
            
            tree[name] = []
            for child in children:
                
                child_name = self.find_name(child)
                tree[name].append(child_name)
                self.construct_tree(child, tree = tree, urls = urls)
                
        return tree, urls
       
    
    
    
    def collect_body_row_information(self, url):
        
        request = requests.get(url)
        information = str(request.content)
        information = information.replace('\\n', '')
        information = information.replace('\\r', '')
        information = information.replace('\\t', '')
        information = information.replace(' /', '')
        information = information.replace('    ', '')
        information = information.replace('\\xc2\\xb0', '')
        information = self.find_between_tags(information, 'div', 'class="col-lg-8"')[0]
        information = self.find_between_tags(information, 'div', 'class="row"')
    
        Data = {}
    
        for item in information:
            
            if len(self.find_between_tags(item, 'div', 'class="col-lg-4"')) != 0:
                
                parameter = self.find_between_tags(item, 'div', 'class="col-lg-4"')[0]
                parameter = parameter[0:-1]
                value = self.find_between_tags(item, 'div', 'class="col-lg-8"')[0]
                
                Data[parameter] = value
                
        Result = {}
                
        for key in Data.keys():
            
            
            if key == 'Type':
                Result['Type'] = self.find_between_tags(Data[key], 'strong')[0]
                surface = self.find_between_tags(Data[key], 'em')
                if len(surface) != 0:
                    Result['Surface'] = surface[0]
                    
                    
            elif key == 'Full Spectral Class':
                Result[key] = self.find_between_tags(Data[key], 'em')[0]       
                Result[key] += ' ' + self.find_between_tags(Data['Luminosity'], 'strong')[0]
                
                
            elif key == 'First scanned for EDSM by':
                try:
                    Result['First scanned'] = self.find_between_tags(Data[key], 'span')[0]
                except: 
                    try:
                        Result['First scanned'] = self.find_between_tags(Data[key], 'a')[0]
                    except:
                        pass
                
                
            elif key == 'First mapped for EDSM by':
                try:
                    Result['First mapped'] = self.find_between_tags(Data[key], 'span')[0]
                except:
                    try:
                        Result['First mapped'] = self.find_between_tags(Data[key], 'a')[0]
                    except:
                        pass
                
                
            elif key == 'Solid composition' or key == 'Atmosphere composition':
                info = Data[key].split('<br>')
                info.pop(-1)
                for item in info:
                    p = item.split('%')
                    Result[p[1]] = float(p[0]) / 100
            
            
            elif key == 'Rotational period':
                Result['Rotational period'] = self.find_between_tags(Data[key], 'strong')[0]
                locked = self.find_between_tags(Data[key], 'em')
                if len(locked) != 0:
                    Result['Tidally Locked'] = True
                else:
                    Result['Tidally Locked'] = False
                    
                    
            elif key == 'Materials':
                materials_names = self.find_between_tags(Data[key], 'strong')
                materials_composition = self.find_between_tags(Data[key], 'small')
                badges = self.find_attribute(Data[key], 'class')
                for index in range(len(materials_names)):
                    Result[materials_names[index]] = materials_composition[index]
                    Result[materials_names[index] + ' badge'] = badges[index].replace('badge badge-material badge-', '')
                    
                    
            else:
                info = self.find_between_tags(Data[key], 'strong')
                if len(info) != 0:
                    Result[key] = info[0]
                    
                    
        
        return Result
        
        
        
        
    def proced_all_information(self):
        
        self.get_information()
        self.construct_tree(self.system_information)
            
        Result = {}
            
        for key in self.urls.keys():
            
            try:
                Result[key] = self.collect_body_row_information(self.urls[key])
            except:
                print('Error acured')
                print('Body name', key)
                print('Body url', self.urls[key])
            else:
                Result[key]['url'] = self.urls[key]
            
        self.system_information = Result
        
        
        
    
    def full(self):
        S = self.system_information
        for key in S.keys():
            print('\t\t', key)
            print()
            for k in S[key].keys():
                print(k, ' ' * (28 - len(k)), S[key][k])
            print('-'*120)
            
    
    
    
    def mini(self):
        print(self.system_name, self.system_id)
        for key in self.tree.keys():
            print(key, end = ' ' * (30 - len(key)))
            print(self.tree[key])
    
i = 13
i = 5
i = 22
A = System_constructor((Systems_data_frame.loc[i])['name'])
A.get_information()
```
